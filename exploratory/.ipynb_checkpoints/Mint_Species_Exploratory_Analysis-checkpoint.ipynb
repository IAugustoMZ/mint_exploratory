```python
# performing data transformation in oil_yield column
data['oil_yield'] = data['oil_yield']*100
```

```python
# function for checking for outliers
def check_outliers(data, var):
    
    # definition of standard colors
    colors = ['green', 'skyblue', 'violet']
    
    # identification of categories of species
    cat_species = data['species'].unique()
    
    # definition of labels
    labels = []
    for i in range(len(cat_species)):
        labels.append('M. %s - Mean: %.2f'%(cat_species[i], 
                                          data.loc[data['species'] == cat_species[i], var].mean()))
    
    # figure creation
    fig = plt.figure(figsize = (15,7))
    ax1 = fig.add_subplot(1,2,1)
    
    # first plot - plotting histogram by species
    for i in range(len(cat_species)):
        ax1.hist(data.loc[data['species'] == cat_species[i], var], color = colors[i], alpha = 0.5,
                label = labels[i])
        
    # first plot configurations
    ax1.set_xlabel(var, size = 16)
    ax1.set_ylabel('Frequency', size = 16)
    ax1.grid()
    ax1.legend()
    ax1.set_title('Distributions of %s \nby Species'%var, size = 18)
    
    # second plot - boxplot by species
    ax2 = fig.add_subplot(1,2,2)
    
    # creating customized color palette and plotting 
    sns.set_palette(sns.color_palette(colors))
    sns.boxplot(data = data, x = 'species', y = var)
    
    # second plot configurations
    ax2.set_xlabel('Mint Species', size = 16)
    ax2.set_ylabel(var, size = 16)
    ax2.grid()
    ax2.set_title('Boxplot of %s \nby Species'%var, size = 18)
    
    # finishing plot configurations
    plt.suptitle('Outlier Checking for %s'%var, size = 24)
    plt.subplots_adjust(top = 0.83)
    
```

```python
# selecting the desired column to check for outliers
select_cols = data.columns.to_list()
to_drop = ['sample_date', 'species', 'sample_no']
select_cols = [x for x in select_cols if x not in to_drop]

# looping over selected columns to perform exploratory analysis
for col in select_cols:
    check_outliers(data, col)
```
